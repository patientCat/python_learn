```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""
利用y = w * x的模型去回归data.txt
使用穷举法，穷举w
"""
```

```python
input = np.loadtxt("data.txt",delimiter=',')
print(input)
```

```python
# 取第一列
x_data = input[:, 0:1]
# 取最后一列
y_data = input[:, 1:]
n_row = input.shape[0]

print("x_data",x_data)
print("y_data",y_data)
```

```python
# 以 y = w * x作为模型进行穷举法训练
def forward(w, x):
    return w * x

def loss(y_hat, y):
    return (y_hat - y) * (y_hat - y)
```

```python
# 穷举0-4
w_range = np.arange(0, 4.1, 0.1)
print(w_range)
print(w_range.size)
```

```python
loss_sum = np.zeros(w_range.size)
for index, w_var in enumerate(w_range):
    loss_v = 0
    for i in range(n_row):
        y_hat = forward(w_var, x_data[i])
        loss_val = loss(y_hat, y_data[i])
        loss_v = loss_v + loss_val
    loss_sum[index] = loss_v
```

```python
plt.plot(w_range, loss_sum)
plt.xlabel("w")
plt.ylabel("loss")
```
